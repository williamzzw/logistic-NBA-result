```python
def get_elo(team):
    try:
        return team_elo[team]
    except:
        return first_elo
```

```python
def update_elo(wteam,lteam):
    win_team_rank=get_elo(wteam)
    lose_team_rank=get_elo(lteam)
    Ewin=1/(1+math.pow(10,(lose_team_rank-win_team_rank)/400))
    if win_team_rank<2100:
        win_score=win_team_rank+32*(1-Ewin)
        diff=win_score-win_team_rank
        lose_score=lose_team_rank-diff
    elif (win_team_rank>=2100)&(win_team_rank<2400):
        win_score=win_team_rank+24*(1-Ewin)
        diff=win_score-win_team_rank
        lose_score=lose_team_rank-diff
    else:
        win_score=win_team_rank+16*(1-Ewin)
        diff=win_score-win_team_rank
        lose_score=lose_team_rank-diff
    team_elo[wteam]=win_score
    team_elo[lteam]=lose_score 
```

```python
def bulid_dataset():
    result=pd.read_csv('18-19first_result.csv')
    x=[]
    y=[]
    for i,row in result.iterrows():
        update_elo(row['WTeam'],row['LTeam'])
        wteam_feature=team_feature.loc[row['WTeam'],:]
        lteam_feature=team_feature.loc[row['LTeam'],:]
        if row['WLoc']=='H':
            wteam_elo=get_elo(row['WTeam'])+100
            lteam_elo=get_elo(row['LTeam'])
        else:
            lteam_elo=get_elo(row['LTeam'])+100
            wteam_elo=get_elo(row['WTeam'])
        wteam_features=wteam_feature.values.tolist()
        wteam_features.append(wteam_elo)
        lteam_features=lteam_feature.values.tolist()
        lteam_features.append(lteam_elo)
        if random.random()>0.5:
            x.append(wteam_features+lteam_features)
            y.append(0)
        else:
            x.append(lteam_features+wteam_features)
            y.append(1)
    return pd.DataFrame(x),pd.Series(y)
```

```python
schedule=pd.read_csv('18-19schedule.csv')
```

```python
def predictwinner(hteam,vteam):
    x=[]
    for k,rows in schedule.iterrows():
        hteam_score=get_elo(hteam)+100
        vteam_score=get_elo(vteam)
        hteam_feature=team_feature.loc[hteam,:].values.tolist()
        vteam_feature=team_feature.loc[vteam,:].values.tolist()
        hteam_feature.append(hteam_score)
        vteam_feature.append(vteam_score)
        x.append(hteam_feature+vteam_feature)
        return pd.DataFrame(x)
    
```

```python
win_team=[]
lose_team=[]
win_proba=[]
for num,row in schedule.iterrows():
    data=predictwinner(row['Hteam'],row['Vteam'])
    schedule_pred=logit_model.predict_proba(data)[:,0]
    if schedule_pred>0.5:
        win_team.append(row['Hteam'])
        lose_team.append(row['Vteam'])
        win_proba.append(schedule_pred)
    else:
        win_team.append(row['Vteam'])
        lose_team.append(row['Hteam'])
        win_proba.append(1-schedule_pred)
        
```

```python
predict_result=pd.DataFrame({'Win':win_team,'lose':lose_team,'proba':win_proba})
```
